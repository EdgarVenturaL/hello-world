```python
import numpy as np
from scipy.integrate import quad

# Función para calcular el momento flector
def moment_at(x, loads, a, b):
    moment = 0
    for load in loads:
        if load[0] < x:
            moment += load[1] * (x - load[0])
        else:
            moment += load[1] * (x - load[0]) / (b - a)
    return moment

# Función para calcular el esfuerzo máximo
def max_stress(loads, a, b):
    def integrand(x):
        return abs(moment_at(x, loads, a, b))

    result, _ = quad(integrand, a, b)
    return result

#  cargas (posición, magnitud)
loads = [(0, 10), (2, 20), (4, 15)] 
a = 0  
b = 5  

max_stress_value = max_stress(loads, a, b)
print("El esfuerzo máximo en la viga es:", max_stress_value)
```
