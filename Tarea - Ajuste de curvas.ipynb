## Primero, importamos los datos usando la función read_csv de Pandas:

```python
import pandas as pd
```

```python
data = pd.read_csv('population_data.csv')
```

```python
import numpy as np

# Datos
x = data.iloc[:, 0]
y = data.iloc[:, 1]

# Ajuste polinomial
p1 = np.polyfit(x, y, 1)
p2 = np.polyfit(x, y, 2)
p3 = np.polyfit(x, y, 3)
```

```python
import matplotlib.pyplot as plt

# Graficar datos y curvas ajustadas
plt.plot(x, y, '.', label='Datos')
plt.plot(x, np.polyval(p1, x), '-', label='Grado 1')
plt.plot(x, np.polyval(p2, x), '--', label='Grado 2')
plt.plot(x, np.polyval(p3, x), '-.', label='Grado 3')
plt.legend()
plt.show()
```

```python
# Calcular errores cuadráticos
e1 = np.sum((np.polyval(p1, x) - y) ** 2)
e2 = np.sum((np.polyval(p2, x) - y) ** 2)
e3 = np.sum((np.polyval(p3, x) - y) ** 2)

# Graficar error acumulado
plt.plot([1, 2, 3], [e1, e2, e3], '-o')
plt.xlabel('Grado del polinomio')
plt.ylabel('Error cuadrático acumulado')
plt.show()
```
